```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# Generate a synthetic dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.0)

# Supervised Learning: Train a classifier on labeled data
svc = SVC(kernel='linear', C=1).fit(X, y)

# Plotting
plt.figure(figsize=(8, 6))

# Scatter plot of data points colored by class
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=100, marker='o')

# Plot decision boundary and decision regions
xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 200), 
                     np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 200))
Z = svc.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap='coolwarm', alpha=0.3)
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')

# Adding legend with class labels
plt.legend(handles=scatter.legend_elements()[0], labels=['Class 1', 'Class 2'], loc='upper right', fontsize=12)

# Title and labels
plt.title("Supervised Learning", fontsize=16)
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)

plt.tight_layout()
plt.show()
```
